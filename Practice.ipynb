```python


spark.conf.set(
  "fs.azure.account.key.sdsource.blob.core.windows.net",
  "<secrete key>" )
```

```python
spark.conf.set(
  "fs.azure.account.key.sdsink.blob.core.windows.net",
  "<secrete key>")
```

```python
src_path ="wasbs://input@sdsource.blob.core.windows.net"
sink_path ="wasbs://output@sdsink.blob.core.windows.net"
df = spark.read.csv(src_path,header=True,inferSchema=True)
display(df)
```

```python
df2 = df.write.mode("overwrite").csv(sink_path)
display(df2)
```
