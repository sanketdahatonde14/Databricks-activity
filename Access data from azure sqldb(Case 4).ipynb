```python
dbutils.secrets.listScopes()
```

```python
dbutils.secrets.list("SanketScope")
```

```python
db_user = dbutils.secrets.get(scope = "SanketScope", key = "dbuser")
db_pass = dbutils.secrets.get(scope = "SanketScope", key = "dbpassword")
display(db_user)
```

```python
#configure a connection to SQL server
jdbcHostname = "sqlserver.database.windows.net"
jdbcDatabase = "AdventureWorks"
jdbcPort = 1433
jdbcUrl = f"jdbc:sqlserver://{jdbcHostname}:{jdbcPort};database={jdbcDatabase}"
connectionProperties = {
  "user" : "sqlserver"      ,
  "password" : dbutils.secrets.get(scope = "SanketScope", key = "sqlserverPassword")                                                                                                                                                
```

```python
df_employees = (
    spark.read
        .format("sqlserver")
        .option("host", "chandewarserver.database.windows.net")
        .option("port", "1433")
        .option("user", db_user)
        .option("password", db_pass)
        .option("database", "SourceTargetDB")
        .option("dbtable", "employee")
        .load()
)

display(df_employees)
```

```python
df_employees.write.format("sqlserver").option("host", "chandewarserver.database.windows.net").option("port", "1433").option("user", db_user).option("password", db_pass).option("database", "SourceTargetDB").option("dbtable", "employees").save()
```
